```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

# 1. Работа с таблицей

## 1.1 Чтение таблицы

```python
directory = "/home/ajumageldiyeva/homework/task2/TRB_10_7"
filename = "TRB_10_7.tsv"
path = os.path.join(directory, filename)
df = pd.read_csv(path, sep='\t')
df.head
```

## 1.2 Построение графика

```python
trbv_counts = df["TRBV7-9"].value_counts()

plt.figure(figsize=(12, 6))
trbv_counts.plot(kind="bar", color="red", edgecolor="black")
plt.xlabel("Тип TRBV")
plt.ylabel("Частота встречаемости")
plt.title("Распределение TRBV в данных")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
